```python
"""The reduce(fun,seq) function is used to apply a particular function passed in its argument to all of the list elements mentioned in the sequence passed along.This function is defined in “functools” module.

Working :  

1. At first step, first two elements of sequence are picked and the result is obtained.
2. Next step is to apply the same function to the previously attained result and the number just succeeding 
   the second element and the result is again stored.
3. This process continues till no more elements are left in the container.
4. The final returned result is returned and printed on console."""

"""The functools module is part of Python's standard library and was implemented for higher-order functions. 
    Higher-order functions are those functions that either accept a function as an input or return 
    another function as an output. For example, map(), filter(), etc. are the higher order functions."""
```

```python
# Combining all dataframe using reduce, merge:

data_frames = [df_2002, df_2003, df_2004, df_2005, df_2006, df_2007, df_2008, df_2009, df_2010,
                      df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, 
                      df_2021, df_2022, df_2023]

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['EN_country'],
                                            how='outer'), data_frames)

df_merged.head()
```

### Note:

Now our basic plots is completed, now what we can do is try to use statistics to answer various conceptual questions like 
"which country most consistently at the highest / lowest rank", avg rank of each country etc. using mean, median, mode and s.d

##### Mean: 
The mean, also known as the average, is calculated by summing up all the values in a dataset and dividing by the total number of values. It represents the central tendency of the data and is highly influenced by outliers. The mean is commonly used to provide a general sense of the data's average value.

##### Median: 
The median is the middle value in a sorted dataset when the values are arranged in ascending or descending order. If the dataset has an odd number of values, the median is the middle value. If the dataset has an even number of values, the median is the average of the two middle values. The median is a robust measure that is not affected by extreme values or outliers, making it useful when dealing with skewed distributions.

##### Mode: 
The mode is the value that appears most frequently in a dataset. It represents the most common or frequently occurring value. A dataset can have one mode (unimodal), two modes (bimodal), or more than two modes (multimodal). The mode is particularly useful for categorical or discrete data, but it can also be applied to numerical data.

##### Standard Deviation: 
The standard deviation is a measure of the dispersion or variability of the data points from the mean. It quantifies how much the individual data points deviate from the average. A higher standard deviation indicates a wider spread of values, while a lower standard deviation suggests that the data points are closer to the mean. The standard deviation is commonly used to understand the distribution and variability of the data.
